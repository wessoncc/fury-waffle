```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
uberfile = "New-York-Transportation/Raw Data/Uber_15-Aug_18.csv"
uber_data = pd.read_csv(uberfile)
uber_data = uber_data[uber_data['Years']!=2018]
#uber_data
```

```python
uber_data15 = uber_data[uber_data['Years']==2015]; uber_data15.reset_index(drop=True)
uber_data16 = uber_data[uber_data['Years']==2016]; uber_data16.reset_index(drop=True)
uber_data17 = uber_data[uber_data['Years']==2017]; uber_data17.reset_index(drop=True)
uber_data17 = uber_data17[uber_data17['Week Number']<14]
#uber_data17
```

```python
target_uber_data = uber_data15.append(uber_data16).append(uber_data17)
target_uber_data = target_uber_data.drop(['DBA'], axis=1)
target_uber_data

#target_uber_data.reset_index(drop=True)
```
